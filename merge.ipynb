```python
import pandas as pd
```

```python
dict1 = {'name':['홍길동', '홍길동', '김철수', '박영희', '김철수', '김철수', '홍길동'], 'korean': range(7)}
df1 = pd.DataFrame(dict1)
```

```python
df1
```

```python
# 색인을 이용하여 병합하기 set_index() 활용 
```

```python
newdf1 = df1.set_index('name')
```

```python
newdf1
```

```python
dict2 = {'name':['김철수', '홍길동', '심수봉'], 'english':range(3)}
df2 = pd.DataFrame(dict2)
```

```python
df2
```

```python
pd.merge(df1, df2, on='name')
```

두 데이터셋 간의 중복된 요소만을 Join하는 Inner Join과 두 데이터셋에 대한 모든 데이터를 Join하는 Outer Join, 그리고 왼쪽 데이터셋을 기준으로 하는 Left Join, 오른쪽 데이터셋을 기준으로 하는 Right Join이 있음. 방식의 차이가 있는 것이지 두 데이터셋을 병합한다는 것은 같음!!
 병합하니까 name은 모두 나열되서 보이고, 칼럼 별 해당 값은 기준이 무엇이냐에 따라서 달라짐. 
 공통적인 이름의 칼럼이 존재해야 해당 칼럼을 이용하여 데이터를 합침. (on='name')

```python
pd.merge(df1, df2, how='inner') # default값은 how='inner'
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

양쪽 데이터 프레임에 동일 이름의 칼럼이 존재하지 않는 경우에는 on 매개변수를 사용할 수 없음. left_on, right_on 사용하기!!

```python
dict3 = {'leftkey':['홍길동', '홍길동', '김철수', '박영희', '김철수', '김철수', '홍길동'], 'korean':range(7)}
```

```python
df3 = pd.DataFrame(dict3)
```

```python
df3
```

```python
dict4 = {'rightkey':['김철수', '홍길동', '심수봉'], 'english': range(3)}
df4 = pd.DataFrame(dict4)
```

```python
df4
```

df3, df4에는 공통되는 이름의 칼럼이 없음.

```python
pd.merge(df3, df4, left_on='leftkey', right_on='rightkey') # leftkey와 rightkey 칼럼에서 비교하여 공통되는 이름만 모아서 병합
```

여러 개의 키를 병합할 때는 칼럼 이름이 들어간 리스트를 넘겨준다. 조인할 칼럼 2개를 리스트로 만들어 이것을 이용하여 병합

```python
dict1 = {'key1':['김철수', '김철수', '박영희'],
          'key2':['one', 'two', 'one'],
         'leftval':[1,2,3]}
left = pd.DataFrame(dict1)
left
```

```python
dict2 = {'key1':['김철수', '김철수', '박영희', '박영희'],
        'key2':['one', 'one', 'one', 'two'],
        'leftval':[4,5,6,7]}
right = pd.DataFrame(dict2)
right
```

```python
mylist = ['key1', 'key2']
pd.merge(left, right, on=mylist, how='outer')
```

```python
pd.merge(left, right, on='key1', suffixes=('_왼쪽', '_오른쪽'))
```
